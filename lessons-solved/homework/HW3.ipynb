```python
import pandas as pd
anscombe_df = pd.read_json('/content/sample_data/anscombe.json')
grouped_df = anscombe_df.groupby('Series')
grouped_df.describe()
#boxplot
grouped_df.boxplot(rot=45, fontsize=12, figsize=(8, 10))

```

```python
#scatterplot
series_values = anscombe_df['Series'].unique()
for i in series_values:
  anscombe_subset = anscombe_df[anscombe_df['Series'] == i]
  anscombe_subset.plot(kind='scatter', x="X", y="Y", title=f'Series {i}')
```
