```python
import pandas as pd
import matplotlib.pyplot as plt
anscombe_df = pd.read_json('/content/sample_data/anscombe.json')
grouped_df = anscombe_df.groupby('Series')
grouped_df.describe()
#boxplot
grouped_df.boxplot(rot=45, fontsize=12, figsize=(8, 10))

#scatterplot
color = ['r', 'b', 'g', 'y']

scatterplot_grouped = plt.subplots()

for i in range(len(color)):
  plot = grouped_df.plot(kind='scatter', x='X', y='Y', color=color[i], title = 'Anscombe Data', subplots=True, legend='top right')

```
